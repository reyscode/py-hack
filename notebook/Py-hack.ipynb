## Details of AE patients

```python
str1='https://pyhack-dot-pharmanlp-177020.uc.r.appspot.com/api/1/StudyHack/ae/subject/'
str2='/list'
```

```python
pat_details = []
for i in list_of_pat['data']:
    temp=requests.get(str1+str(i)+str2)
    pat_details.append(temp)
    print(i, len(pat_details), '/', len(list_of_pat['data']))
    if 'data' not in temp.json():
        break
```

```python
pdetails = []
for patient in pat_details:
    details = patient.json()['data']
    pdetails.append(details)
```
